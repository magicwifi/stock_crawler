```python
# coding:utf-8
import arrow
import qgrid
import seaborn as sns
import numpy as np
import pandas as pd
from pandas import Series, DataFrame

import matplotlib.pyplot as plt
%matplotlib inline

sns.set(style='whitegrid')
```

```python
trans_prov_allcount = pd.read_csv("../a_stock_news/result.csv",encoding='utf-8',index_col = False, delimiter=',' )
trans_prov_allcount.head()
```

```python
import re

def contail_char(x):
    if len(re.findall('[a-zA-Z0-9]+',str(x)))==0:
        return True;
    else:
        return False;
    
#result_list = re.findall('[a-zA-Z0-9]+',input_line)

trans_prov_allcount = trans_prov_allcount[trans_prov_allcount['company'].apply(contail_char)]

```

```python
def title_contain(x):
    if str(x['company']) in str(x['title']):
        return True;
    else:
        return False

trans_prov_allcount = trans_prov_allcount[trans_prov_allcount.apply(title_contain,axis=1)]
trans_prov_allcount.head()
```

```python
trans_prov_allcount= trans_prov_allcount.drop_duplicates(subset='title', keep='last')
trans_prov_allcount.shape
```

```python
trans_prov_allcount = trans_prov_allcount.drop(['ctime','label_type','label_article','industry','url','label_property','code'],axis=1)
trans_prov_allcount.head()
```

```python
company_list = trans_prov_allcount['company'].unique()
```

```python
with open('a_company.csv', 'w') as f:
    for item in company_list:
        f.write("%s\n" % item)
```
